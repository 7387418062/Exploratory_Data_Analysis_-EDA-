```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
data = {
    'StudentID': range(1, 101),
    'Subject1': np.random.randint(0, 100, size=100),
    'Subject2': np.random.randint(0, 100, size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100)
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Sample of the dataset:")
print(df.head())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nData Types and Missing Values:")
print(df.info())
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(2, 2, 1)
sns.histplot(df['Subject1'], kde=True)
plt.title('Subject1 Scores Distribution')
```

```python
plt.subplot(2, 2, 2)
sns.histplot(df['Subject2'], kde=True)
plt.title('Subject2 Scores Distribution')
```

```python
plt.subplot(2, 2, 3)
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
```

```python
plt.subplot(2, 2, 4)
sns.scatterplot(x='Subject1', y='Subject2', data=df)
plt.title('Scatter Plot: Subject1 vs Subject2 Scores')
```

```python
df.loc[df.sample(frac=0.1).index, 'Subject1'] = np.nan
```

```python
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
df_cleaned = df.dropna()
```

```python
# Display plots
plt.tight_layout()
plt.show()
```
